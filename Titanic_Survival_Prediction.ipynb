```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
df.head()
```

```python
df.info()
```

```python
print(df.shape)
df.isnull().sum()
```

```python
# majority missing values in 'cabin' so drop column
df = df.drop(columns='Cabin',axis=1)
df.head()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
print(df['Embarked'].mode()[0])
```

```python
# replacing missing embark values with mode value
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# finding the number of people survived
df['Survived'].value_counts()
```

```python
import seaborn as sns

sns.set()
# 1. survival count
sns.countplot(x='Survived', data=df, palette=['skyblue', 'salmon'])
plt.title('Count of survived vs Not Survived')
plt.show()
```

```python
# 2. Survival Count by Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df, palette=['skyblue', 'salmon'])
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# 2. number of survivals gender wise
sns.countplot(x='Sex',hue = 'Survived', data = df)
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)

# Add a title to the plot
plt.title('Number of Survivors by Passenger Class')
```

```python
#Encoding categorical column
df['Sex'].value_counts()
```

```python
df['Embarked'].value_counts()
```

```python
# converting categorical column
df.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}}, inplace = True)
```

```python
df.head()
```

```python
X = df.drop(columns = ['PassengerId','Name','Ticket','Survived'], axis=1)
Y = df['Survived']
```

```python
print("X=\n",X,"Y=\n",Y)
```

```python
#splitting data into training and testing
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Model training
# As we want to predict survival in 0 or 1 we'll using
#LOGISTIC REGRESION MODEL
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# Predict on the train data
X_train_pred = model.predict(X_train)

# Calculate and print the accuracy
accuracy = accuracy_score(Y_train, X_train_pred)

# Define bold text formatting
bold = '\033[1m'
reset = '\033[0m'

# Print the accuracy score in bold
print(f"{bold}Accuracy score : {accuracy}{reset}")
```

```python
# Predict on the test data
Y_pred = model.predict(X_test)

# Calculate and print the accuracy
accuracy2 = accuracy_score(Y_test, Y_pred)

# Define bold text formatting
bold = '\033[1m'
reset = '\033[0m'

# Print the accuracy score in bold
print(f"{bold}Accuracy score : {accuracy2}{reset}")
```
